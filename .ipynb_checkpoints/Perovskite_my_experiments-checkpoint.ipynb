Well, not only these... we need other features for better analysis. We can extract crystal system, formula, seperate cations and anions maybe... these kind of manipulations are to be done!

```python
# Querying for structure data, crystal system and elements..
data2 = mpr.materials.summary.search(formula='ABX3', fields=['structure','is_metal','symmetry'])
```

```python
data2[0]
```

```python
data_dicts = [doc.dict() for doc in data2] # converting MPDataDoc<SummaryDoc> to dictonary
```

```python
perovskite_df = pd.DataFrame(data_dicts)
```

```python
columns_of_interest_2 = ['symmetry','structure','is_metal']
refined_perovskite_df_2 = perovskite_df[columns_of_interest_2]
```

```python
refined_perovskite_df_2
```

```python
# Extracting crystal_system

refined_perovskite_df_2['crystal_system'] = refined_perovskite_df_2['symmetry'].apply(lambda x: x['crystal_system'])
```

```python
refined_perovskite_df_2
```

```python
# Extracting abc and angles

def extract_abc_angles(structure):
    abc = structure['lattice']['a'], structure['lattice']['b'], structure['lattice']['c']
    angles = structure['lattice']['alpha'], structure['lattice']['beta'], structure['lattice']['gamma']
    return pd.Series({'a': abc[0], 'b': abc[1], 'c': abc[2], 'alpha': angles[0], 'beta': angles[1], 'gamma': angles[2]})

# Convert the structure column to dictionary if it's a string
# refined_perovskite_df_2['structure'] = refined_perovskite_df_2['structure'].apply(eval)

# Apply the function to extract values and create new columns in the dataframe
refined_perovskite_df_2[['a', 'b', 'c', 'alpha', 'beta', 'gamma']] = refined_perovskite_df_2['structure'].apply(extract_abc_angles)
```

```python
refined_perovskite_df_2
```

```python
refined_perovskite_df_2 = refined_perovskite_df_2.drop(columns=['symmetry','structure'])
```

```python
refined_perovskite_df_2
```

```python
# Dropping the metal rows
refined_perovskite_df_2 = refined_perovskite_df_2[refined_perovskite_df_2.is_metal == False].copy()
```

```python
refined_perovskite_df_2
```

```python
refined_df.info()
```

```python
# Merging the two data frames
merged_refined_df = pd.merge(refined_df,refined_perovskite_df_2, on='Index') 
```
