```python
# We'll get the ionic radii now
import ast
# Function to parse strings into dictionaries (if necessary)
def parse_dict_string(ionic_radius):
    if isinstance(ionic_radius, str):
        return ast.literal_eval(ionic_radius)  # Convert string to dictionary
    return ionic_radius  # Already a dictionary, return as is

# Apply the parsing function to all the ionic radius columns
df['element_1_ionic_radius'] = df['element_1_ionic_radius'].apply(parse_dict_string)
df['element_2_ionic_radius'] = df['element_2_ionic_radius'].apply(parse_dict_string)
df['element_3_ionic_radius'] = df['element_3_ionic_radius'].apply(parse_dict_string)

# Function to get ionic radius for +3 for element 1 and element 2, and -2 for element 3
def get_ionic_radius(ionic_radius_dict, default_charge):
    if ionic_radius_dict and isinstance(ionic_radius_dict, dict):
        try:
            # Try to get the radius for the given charge (e.g., +3 for element 1 and 2, -2 for element 3)
            return ionic_radius_dict[str(default_charge)]
        except KeyError:
            # If the desired charge is not available, return the first available one if it exists
            if ionic_radius_dict:
                return list(ionic_radius_dict.values())[0]
            else:
                return None  # No available values in the dictionary
    return None  # Handle cases where ionic_radius_dict is empty or not a dictionary

# Apply the function to extract ionic radii
df['element_1_ionic_radius_final'] = df['element_1_ionic_radius'].apply(lambda x: get_ionic_radius(x, 3))
df['element_2_ionic_radius_final'] = df['element_2_ionic_radius'].apply(lambda x: get_ionic_radius(x, 3))
df['element_3_ionic_radius_final'] = df['element_3_ionic_radius'].apply(lambda x: get_ionic_radius(x, -2))

# Drop the original ionic radius columns if needed
df = df.drop(columns=['element_1_ionic_radius', 'element_2_ionic_radius', 'element_3_ionic_radius'])
```

```python
def calculate_weighted_stats(df):
    """
    Calculate weighted mean and standard deviation for atomic numbers.
    
    Parameters:
    df (DataFrame): DataFrame containing at_num1, at_num2, at_num3 columns.
    
    Returns:
    DataFrame: Original DataFrame with additional columns for mean and std.
    """
    # Calculate the weighted mean
    df['mean_at_num'] = (df['atomic_num1'] + df['atomic_num2'] + 3 * df['atomic_num3']) / 5
    
    # Calculate the weighted standard deviation
    mean_value = df['mean_at_num']
    df['std_at_num'] = np.sqrt(
        (1/5) * (
            (df['atomic_num1'] - mean_value) ** 2 +
            (df['atomic_num2'] - mean_value) ** 2 +
            (3 * (df['atomic_num3'] - mean_value) ** 2)
        )
    )

    return df
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error
from sklearn.decomposition import PCA

# Assuming you have your DataFrame `df` ready and the relevant functions defined

# Define the feature subset for PCA and weighted method
radii_features = [
    'element_1_atomic_radius', 'element_2_atomic_radius', 'element_3_atomic_radius',
    'element_1_ionic_radius_final', 'element_2_ionic_radius_final', 'element_3_ionic_radius_final',
    'element_1_van_der_waals_radius', 'element_2_van_der_waals_radius', 'element_3_van_der_waals_radius'
]

# Split data for training and testing
X_train, X_test, y_train, y_test = train_test_split(df[radii_features], df['band_gap'], test_size=0.2, random_state=42)

# Function to evaluate model performance
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    return r2, mae

# Initialize models
models = {
    'Linear Regression (PCA)': LinearRegression(),
    'Random Forest': RandomForestRegressor(random_state=42),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42)
}

# Apply PCA
pca = PCA(n_components=0.95)  # Preserve 95% variance
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

# Evaluate PCA for Linear Regression
print("Evaluating Linear Regression with PCA:")
r2_pca_lr, mae_pca_lr = evaluate_model(models['Linear Regression (PCA)'], X_train_pca, y_train, X_test_pca, y_test)
print(f"PCA Model R2: {r2_pca_lr}, MAE: {mae_pca_lr}")

# Evaluate Random Forest and Gradient Boosting without PCA
for model_name, model in models.items():
    if model_name != 'Linear Regression (PCA)':
        print(f"\nEvaluating {model_name}:")
        r2, mae = evaluate_model(model, X_train, y_train, X_test, y_test)
        print(f"{model_name} R2: {r2}, MAE: {mae}")

# Weighted Feature Method
df = calculate_weighted_radii(df)  # Make sure you have this function defined

# Define weighted features
weighted_features = ['mean_atomic_radius', 'std_atomic_radius']  # Adjust if necessary

# Evaluate models with weighted features
for model_name, model in models.items():
    print(f"\nEvaluating {model_name} with Weighted Features:")
    r2, mae = evaluate_model(model, df[weighted_features], df['band_gap'], df[weighted_features], df['band_gap'])
    print(f"{model_name} with Weighted Features R2: {r2}, MAE: {mae}")
```

```python
def calculate_weighted_physical(df):
    # Weighted mean for atomic mass
    df['mean_atomic_mass'] = (
        df['element_1_atomic_mass'] + df['element_2_atomic_mass'] + 3 * df['element_3_atomic_mass']
    ) / 5
    
    # Weighted standard deviation for atomic mass
    mean_value = df['mean_atomic_mass']
    df['std_atomic_mass'] = np.sqrt(
        (1/5) * (
            (df['element_1_atomic_mass'] - mean_value) ** 2 +
            (df['element_2_atomic_mass'] - mean_value) ** 2 +
            3 * (df['element_3_atomic_mass'] - mean_value) ** 2
        )
    )
    
    
    # Weighted mean for density
    df['mean_density'] = (
        df['element_1_density'] + df['element_2_density'] + 3 * df['element_3_density']
    ) / 5
    
    # Weighted standard deviation for density
    mean_value = df['mean_density']
    df['std_density'] = np.sqrt(
        (1/5) * (
            (df['element_1_density'] - mean_value) ** 2 +
            (df['element_2_density'] - mean_value) ** 2 +
            3 * (df['element_3_density'] - mean_value) ** 2
        )
    )
    
    # Weighted mean for molar volume
    df['mean_molar_volume'] = (
        df['element_1_molar_volume'] + df['element_2_molar_volume'] + 3 * df['element_3_molar_volume']
    ) / 5
    
    # Weighted standard deviation for molar volume
    mean_value = df['mean_molar_volume']
    df['std_molar_volume'] = np.sqrt(
        (1/5) * (
            (df['element_1_molar_volume'] - mean_value) ** 2 +
            (df['element_2_molar_volume'] - mean_value) ** 2 +
            3 * (df['element_3_molar_volume'] - mean_value) ** 2
        )
    )

    
    return df
```
