```python
# We'll get the ionic radii now
import ast
# Function to parse strings into dictionaries (if necessary)
def parse_dict_string(ionic_radius):
    if isinstance(ionic_radius, str):
        return ast.literal_eval(ionic_radius)  # Convert string to dictionary
    return ionic_radius  # Already a dictionary, return as is

# Apply the parsing function to all the ionic radius columns
df['element_1_ionic_radius'] = df['element_1_ionic_radius'].apply(parse_dict_string)
df['element_2_ionic_radius'] = df['element_2_ionic_radius'].apply(parse_dict_string)
df['element_3_ionic_radius'] = df['element_3_ionic_radius'].apply(parse_dict_string)

# Function to get ionic radius for +3 for element 1 and element 2, and -2 for element 3
def get_ionic_radius(ionic_radius_dict, default_charge):
    if ionic_radius_dict and isinstance(ionic_radius_dict, dict):
        try:
            # Try to get the radius for the given charge (e.g., +3 for element 1 and 2, -2 for element 3)
            return ionic_radius_dict[str(default_charge)]
        except KeyError:
            # If the desired charge is not available, return the first available one if it exists
            if ionic_radius_dict:
                return list(ionic_radius_dict.values())[0]
            else:
                return None  # No available values in the dictionary
    return None  # Handle cases where ionic_radius_dict is empty or not a dictionary

# Apply the function to extract ionic radii
df['element_1_ionic_radius_final'] = df['element_1_ionic_radius'].apply(lambda x: get_ionic_radius(x, 3))
df['element_2_ionic_radius_final'] = df['element_2_ionic_radius'].apply(lambda x: get_ionic_radius(x, 3))
df['element_3_ionic_radius_final'] = df['element_3_ionic_radius'].apply(lambda x: get_ionic_radius(x, -2))

# Drop the original ionic radius columns if needed
df = df.drop(columns=['element_1_ionic_radius', 'element_2_ionic_radius', 'element_3_ionic_radius'])
```

# We are going to categorize the properties into groups for making our work easier.

#### Crystallographic Properties: 
a, b, c, alpha, beta, gamma, crystal_Cubic, Hexagonal, Monoclinic, Orthorombhic, Tetragonal, Triclinic, Trigonal

```python
# We will make lists of these columns

# Periodic Table Properties
periodic_table_properties = [
    'atomic_num1', 'atomic_num2', 'atomic_num3',
    'element_1_group_number', 'element_2_group_number', 'element_3_group_number',
    'element_1_row_number', 'element_2_row_number', 'element_3_row_number',
    'element_1_mendeleev_no', 'element_2_mendeleev_no', 'element_3_mendeleev_no'
]

# Element Size Properties
element_size_properties = [
    'element_1_atomic_radius', 'element_2_atomic_radius', 'element_3_atomic_radius',
    'element_1_ionic_radius_final', 'element_2_ionic_radius_final', 'element_3_ionic_radius_final',
    'element_1_atomic_radius_calculated', 'element_2_atomic_radius_calculated', 'element_3_atomic_radius_calculated',
    'element_1_avg_ionic_radius', 'element_2_avg_ionic_radius', 'element_3_avg_ionic_radius',
    'element_1_covalent_radius', 'element_2_covalent_radius', 'element_3_covalent_radius',
    'element_1_van_der_waals_radius', 'element_2_van_der_waals_radius', 'element_3_van_der_waals_radius'
]

# Electrical Properties
electrical_properties = [
    'element_1_electron_affinity', 'element_2_electron_affinity', 'element_3_electron_affinity',
    'element_1_electronegativity', 'element_2_electronegativity', 'element_3_electronegativity',
    'element_1_first_ionization_energy', 'element_2_first_ionization_energy', 'element_3_first_ionization_energy',
    'element_1_dipole_polarizability', 'element_2_dipole_polarizability', 'element_3_dipole_polarizability'
]

# Physical Properties
physical_properties = [
    'element_1_atomic_mass', 'element_2_atomic_mass', 'element_3_atomic_mass',
    'element_1_density', 'element_2_density', 'element_3_density',
    'element_1_molar_volume', 'element_2_molar_volume', 'element_3_molar_volume',
    'element_1_density_of_solid','element_2_density_of_solid','element_3_density_of_solid'
]

# Thermal Properties
thermal_properties = [
    'element_1_boiling_point', 'element_2_boiling_point', 'element_3_boiling_point',
    'element_1_melting_point', 'element_2_melting_point', 'element_3_melting_point',
    'element_1_thermal_conductivity', 'element_2_thermal_conductivity', 'element_3_thermal_conductivity',
    'element_1_coefficient_of_linear_thermal_expansion', 'element_2_coefficient_of_linear_thermal_expansion', 
    'element_3_coefficient_of_linear_thermal_expansion', 'element_1_specific_heat', 'element_2_specific_heat', 
    'element_3_specific_heat', 'element_1_fusion_heat', 'element_2_fusion_heat', 'element_3_fusion_heat',
    'element_1_evaporation_heat', 'element_2_evaporation_heat', 'element_3_evaporation_heat'
]

# Crystallographic Properties
crystallographic_properties = [
    'a', 'b', 'c', 'alpha', 'beta', 'gamma',
    'crystal_Cubic', 'crystal_Hexagonal', 'crystal_Monoclinic', 'crystal_Orthorhombic', 
    'crystal_Tetragonal', 'crystal_Triclinic', 'crystal_Trigonal'
]

# Perovskite Physical Properties
perovskite_physical_properties = [
    'volume', 'density', 'density_atomic'
]

# Other Perovskite Properties
other_perovskite_properties = [
    'nsites', 'is_stable', 'is_gap_direct', 'is_magnetic', 'ordering', 
    'total_magnetization', 'total_magnetization_normalized_vol', 'total_magnetization_normalized_formula_units',
    'energy_above_hull', 'energy_per_atom', 'formation_energy_per_atom', 'band_gap', 'efermi'
]

# Combining all the lists into a dictionary for easy reference
feature_categories = {
    "Periodic Table Properties": periodic_table_properties,
    "Element Size Properties": element_size_properties,
    "Electrical Properties": electrical_properties,
    "Physical Properties": physical_properties,
    "Thermal Properties": thermal_properties,
    "Crystallographic Properties": crystallographic_properties,
    "Perovskite Physical Properties": perovskite_physical_properties,
    "Other Perovskite Properties": other_perovskite_properties
}

# Print the dictionary to verify
for category, features in feature_categories.items():
    print(f"{category}: {features}\n")
```
